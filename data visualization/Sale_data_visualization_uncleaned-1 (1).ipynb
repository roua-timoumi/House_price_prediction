```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the raw dataset
file_path = r"combined_data.csv"  
raw_data = pd.read_csv(file_path)
raw_data=raw_data.head(50)
```

```python
# 1. Price Distribution 
plt.figure(figsize=(20, 10))
sns.histplot(raw_data['Price'], kde=False, color='blue', bins=50)
plt.title('Price Distribution (Uncleaned)')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Area (m) Distribution 
plt.figure(figsize=(20,10))
sns.histplot(raw_data['Area (m)'], kde=False, color='green', bins=50)
plt.title('Area (m) Distribution (Uncleaned)')
plt.xlabel('Area (m)')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python

# 3. Scatter Plot: Price vs Area (m)
plt.figure(figsize=(25, 10))
sns.scatterplot(x=raw_data['Area (m)'], y=raw_data['Price'], alpha=0.6)
plt.title('Price vs Area (m) (Uncleaned)')
plt.xlabel('Area (m)')
plt.ylabel('Price')
plt.show()
```

```python

plt.figure(figsize=(20, 10))
sns.scatterplot(x='Bedrooms', y='Price', data=raw_data, alpha=0.6, hue='Bedrooms', palette='viridis')
plt.title('Scatter Plot: Price vs. Number of Bedrooms (Uncleaned)')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()
```
