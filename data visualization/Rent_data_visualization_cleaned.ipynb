```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r"cleaned_final_version.csv"  
raw_data = pd.read_csv(file_path)
raw_data=raw_data.iloc[560:600]
```

```python
# 1. Price Distribution 
plt.figure(figsize=(8, 5))
sns.histplot(raw_data['Price'], kde=False, color='blue', bins=50)
plt.title('Price Distribution (cleaned)')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Area (m) Distribution 
plt.figure(figsize=(8, 5))
sns.histplot(raw_data['Area (m)'], kde=False, color='green', bins=50)
plt.title('Area (m) Distribution (cleaned)')
plt.xlabel('Area (m)')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
raw_data = pd.read_csv(file_path)

# 3. Scatter Plot: Price vs Area (m) 
plt.figure(figsize=(8, 5))
sns.scatterplot(x=raw_data['Area (m)'], y=raw_data['Price'], alpha=0.6)
plt.title('Price vs Area (m) (cleaned)')
plt.xlabel('Area (m)')
plt.ylabel('Price')
plt.show()
```

```python
raw_data = pd.read_csv(file_path)

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Bedrooms', y='Price', data=raw_data, alpha=0.6, hue='Bedrooms', palette='viridis')
plt.title('Scatter Plot: Price vs. Number of Bedrooms (cleaned)')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()
```

```python
# Visualiser la variation des prix par Governorate
raw_data = pd.read_csv(file_path)

plt.figure(figsize=(12, 6))
sns.boxplot(x='Property Type', y='Price', data=raw_data, showfliers=False)
plt.title('Variation du Prix par Property type')
plt.xlabel('Property type')
plt.ylabel('Price')
plt.xticks(rotation=45)  
plt.show()
```

```python
# Visualiser la variation des prix par Governorate 
plt.figure(figsize=(12, 6))
sns.boxplot(x='Governorate', y='Price', data=raw_data, showfliers=False)
plt.title('Variation du Prix par Governorate')
plt.xlabel('Governorate')
plt.ylabel('Price')
plt.xticks(rotation=45)  
plt.show()
```

```python

combined_data = raw_data

numeric_columns = ['Area (m)', 'Bedrooms', 'Bathrooms', 'Price']

for col in numeric_columns:
    combined_data[col] = pd.to_numeric(combined_data[col], errors='coerce')  

filtered_data = combined_data[numeric_columns].dropna()

# Calculate the correlation matrix
correlation_matrix = filtered_data.corr()

# Plot the correlation matrix using Seaborn heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title('Correlation Matrix: Area (m), Bedrooms, Bathrooms, Price')
plt.show()
```
